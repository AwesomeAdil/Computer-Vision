```python
import torch
from matplotlib import pyplot as plt
import numpy as np
import cv2 as cv
import pyautogui
screenWidth, screenHeight = 3456, 2234
```

```python
model = torch.hub.load('../yolov5', 'yolov5s', source='local')
```

```python
while True:
    screen = pyautogui.screenshot()
    screen_array = np.array(screen)
    # height, width, channels
    cropped_region = screen_array[:, screenWidth//2:7*screenWidth//8, :]
    #convert to BGR
    corrected_colors = cv.cvtColor(cropped_region, cv.COLOR_RGB2BGR)
    
    results = model(corrected_colors)
    cv.imshow('YOLO', corrected_colors)
    cv.waitKey(1)
#     if cv.waitKey(1) & 0xFF == ord('q'):
#         break
cv.destroyAllWindows()
```

```python
import uuid
import os
import time
```

```python
IMAGES_PATH = os.path.join('data', 'images')
number_imgs = 100
```

```python
for img_num in range(50+number_imgs):
    screen = pyautogui.screenshot()
    screen_array = np.array(screen)
    cropped_region = screen_array[:, screenWidth//2:7*screenWidth//8, :]
    corrected_colors = cv.cvtColor(cropped_region, cv.COLOR_RGB2BGR)
    cv.imshow('YOLO', corrected_colors)
    img_name = os.path.join(IMAGES_PATH, str(uuid.uuid1())+'.jpg')
    if img_num > 49:
        print('Collecting image number {}'.format(img_num-49))
        cv.imwrite(img_name, corrected_colors)
    elif not img_num%10:
        print('Part {}'.format(img_num//10))
    cv.waitKey(1)
    time.sleep(.2)
```
