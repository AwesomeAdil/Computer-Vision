```python
import torch
from matplotlib import pyplot as plt
import numpy as np
import cv2 as cv
import pyautogui
screenWidth, screenHeight = 3456, 2234
guiWidth, guiHeight = pyautogui.size()
guiWidth, guiHeight
```

```python
img = cv.imread("../Desktop/Icons/Message.png") 
corrected_img = cv.cvtColor(img, cv.COLOR_RGB2BGR)
messageLocation = pyautogui.locateOnScreen(corrected_img, confidence=.8)
messageLocationCenter = pyautogui.center(messageLocation)
messagex, messagey = messageLocationCenter.x, messageLocationCenter.y
pyautogui.moveTo(messagex/screenWidth * guiWidth, messagey/screenHeight * guiHeight, 3)
messagex, messagey
```

```python
screen = pyautogui.screenshot()
screen_array = np.array(screen)
cropped_region = screen_array[2090:2200,
                              390: 500, 
                              :]
corrected_colors = cv.cvtColor(cropped_region, cv.COLOR_RGB2BGR)
cv.imshow('YOLO', corrected_colors)
cv.waitKey()
```
